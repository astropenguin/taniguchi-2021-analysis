# Schematic diagram of L+S decomposition

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import uniform_filter, median_filter
from jupyter_io import savefig_in_notebook
from matplotlib.gridspec import GridSpec
```

```python
def gaussian(x, mu, sigma):
    return np.exp(-(x-mu)**2 / (2*sigma**2)) / np.sqrt(2*np.pi*sigma**2)

def remove_borders(ax):
    for spine in ax.spines.values():
        spine.set_visible(False)

def remove_ticks(ax):
    ax.set_xticks([])
    ax.set_yticks([])
```

## Common parameters

```python
D = 200
N = 200

t = np.arange(N)
f = np.linspace(-1, 1, D)
f_fine = np.linspace(-1, 1, 10*D)
```

## Simulation

### Low-rank matrix (L)

```python
np.random.seed(2021)

gain = -1 * (f-0.75) * (f+0.75) * (f-0.5) * (f+0.5)
atm = 0.01 * uniform_filter(np.cumsum(np.random.randn(N)), int(N/20))

L = gain[:, np.newaxis] + atm
L -= np.max(L) - np.ptp(L) / 2
```

### Sparse matrix (S)

```python
phi = np.repeat([0, 1, 0, 1, 0, 1, 0], np.ceil(N/7))[:N]
spec = 0.015 * gaussian(f, 0.25, 0.03) + 0.01 * gaussian(f, -0.5, 0.03)
spec_fine = 0.015 * gaussian(f_fine, 0.25, 0.03) + 0.01 * gaussian(f_fine, -0.5, 0.03)

S = -spec[:, np.newaxis] * phi
```

### Noise matrix (E)

```python
np.random.seed(2021)

E = 0.075 * np.random.randn(D, N)
```

### Observed matrix (X)

```python
X = L + S + E
```

## Schematic diagram of L+S decomposition

```python
fig = plt.figure(figsize=(9, 3), dpi=300)
grid = GridSpec(2, 4, height_ratios=[4, 1], left=0.05, right=0.95, top=0.95, bottom=0.05, hspace=0.25, wspace=0.5)
cmap = "viridis"

ax = fig.add_subplot(grid[0, 0])
im = ax.imshow(X, cmap=cmap)
ax.set_title(r"$X$")
ax.set_xlabel(r"Observed time ($t$)")
ax.set_ylabel(r"Measured frequency ($f$)")
remove_borders(ax)
remove_ticks(ax)

ax = fig.add_subplot(grid[0, 1])
ax.imshow(L, clim=im.get_clim(), cmap=cmap)
ax.set_title(r"$L$ (low-rank)")
ax.set_xlabel(r"Observed time ($t$)")
ax.annotate("=", (-0.25, 0.5), xycoords=ax, size=15, ha="center", va="center")
remove_borders(ax)
remove_ticks(ax)

ax = fig.add_subplot(grid[0, 2])
ax.imshow(S, clim=im.get_clim(), cmap=cmap)
ax.set_title(r"$S$ (sparse)")
ax.set_xlabel(r"Observed time ($t$)")
ax.annotate("+", (-0.25, 0.5), xycoords=ax, size=15, ha="center", va="center")
remove_borders(ax)
remove_ticks(ax)

ax = fig.add_subplot(grid[0, 3])
ax.imshow(E, clim=im.get_clim(), cmap=cmap)
ax.set_title(r"$E$ (noise)")
ax.set_xlabel(r"Observed time ($t$)")
ax.annotate("+", (-0.25, 0.5), xycoords=ax, size=15, ha="center", va="center")
remove_borders(ax)
remove_ticks(ax)

ax = fig.add_subplot(grid[1, 1])
ax.plot(t, atm, color="black")
ax.set_xlim(0, 200)
ax.set_ylabel(r"$\mathrm{PWV}(t)$", rotation=0, labelpad=20)
remove_borders(ax)
remove_ticks(ax)

ax = fig.add_subplot(grid[1, 2])
ax.yaxis.tick_right()
ax.plot(t, phi, color="black")
ax.set_xlim(0, 200)
ax.set_ylabel(r"$\phi(t)$", rotation=0, labelpad=15)
ax.set_xticks([])
ax.set_yticks([0, 1])
ax.tick_params(length=0)
ax.set_yticklabels(["0 (OFF)", "1 (ON)"])
remove_borders(ax)

savefig_in_notebook(fig, "lowrank-sparse-decomposition.pdf")
```

## Schematic diagram of algorithm

```python
k = 30
w = 7

S_est = X - L
s_est = (S_est * phi).sum(1) / phi.sum()
psi = np.zeros_like(s_est)
psi[np.argsort(s_est)[:k]] = 1

s_est_filtered = median_filter(s_est, w)
psi_filtered = np.zeros_like(s_est_filtered)
psi_filtered[np.argsort(s_est_filtered)[:k]] = 1

Omega = phi * psi_filtered[:, np.newaxis]
```

```python
fig = plt.figure(figsize=(9, 3), dpi=300)
grid = GridSpec(2, 6, width_ratios=[4, 1, 1, 1, 4, 4], height_ratios=[3, 1],
                left=0.05, right=0.95, top=0.90, bottom=0.05, hspace=0.5, wspace=0.5)
cmap = "viridis"

ax = fig.add_subplot(grid[0, 0])
ax.imshow(X-L, clim=im.get_clim(), cmap=cmap)
ax.set_title(r"$X-L_{n}$")
ax.set_xlabel(r"Observed time ($t$)")
ax.set_ylabel(r"Measured frequency ($f$)")
remove_borders(ax)
remove_ticks(ax)

ax = fig.add_subplot(grid[0, 1])
ax.plot(-s_est, -f, color="black")
ax.set_ylim(-1, 1)
ax.set_title(r"$s$")
ax.annotate(r"$\rightarrow$", (-0.65, 0.5), xycoords=ax, size=15, ha="center", va="center")
remove_borders(ax)
remove_ticks(ax)

ax = fig.add_subplot(grid[0, 2])
ax.plot(-s_est_filtered, -f, color="black")
ax.set_ylim(-1, 1)
ax.set_title(r"($s_{\mathrm{smooth}}$)")
ax.annotate(r"$\rightarrow$", (-0.65, 0.5), xycoords=ax, size=15, ha="center", va="center")
remove_borders(ax)
remove_ticks(ax)

ax = fig.add_subplot(grid[0, 3])
ax.plot(psi_filtered, -f, color="black")
ax.set_ylim(-1, 1)
ax.set_title(r"$\theta_{k}$")
ax.annotate(r"$\rightarrow$", (-0.65, 0.5), xycoords=ax, size=15, ha="center", va="center")
ax.set_xticks([0, 1])
ax.set_yticks([])
ax.tick_params(length=0)
ax.set_xticklabels(["0", "1"])
remove_borders(ax)

ax = fig.add_subplot(grid[0, 4])
ax.imshow(Omega, cmap="binary_r")
ax.set_title(r"$\Omega_{n} = \theta_{k} \phi^{T}$")
ax.annotate(r"$\rightarrow$", (-0.17, 0.5), xycoords=ax, size=15, ha="center", va="center")
ax.annotate(r"$\uparrow$", (0.5, -0.18), xycoords=ax, size=15, ha="center", va="center", zorder=-100)
remove_borders(ax)
remove_ticks(ax)

ax = fig.add_subplot(grid[0, 5])
ax.imshow(Omega*(X-L), clim=im.get_clim(), cmap=cmap)
ax.set_title(r"$S_{n} = \Omega_{n} \circ (X-L_{n})$")
ax.annotate(r"$\rightarrow$", (-0.17, 0.5), xycoords=ax, size=15, ha="center", va="center")
remove_borders(ax)
remove_ticks(ax)

ax = fig.add_subplot(grid[1, 4], sharex=ax)
ax.yaxis.tick_right()
ax.plot(t, phi, color="black")
ax.set_xlim(0, 200)
ax.set_ylabel(r"$\phi^{T}$", rotation=0, labelpad=15)
ax.set_xticks([])
ax.set_yticks([0, 1])
ax.tick_params(length=0)
ax.set_yticklabels(["0", "1"])
remove_borders(ax)

savefig_in_notebook(fig, "sparseid-for-fast-sampled-psw-observations.pdf")
```

## Mock simulation

```python
is_on = phi.astype(bool)
is_baseline = ~psi_filtered.astype(bool)
scan_id = np.cumsum(np.abs(np.hstack([0, np.diff(phi)])))

spec_psw = -np.vstack([
    X[:,scan_id==1].mean(1) - X[:,scan_id==0].mean(1),
    X[:,scan_id==3].mean(1) - X[:,scan_id==2].mean(1),
    X[:,scan_id==5].mean(1) - X[:,scan_id==4].mean(1),
]).mean(0)

spec_godec = -(X-L)[:,is_on].mean(1)

spec_psw -= spec_psw[is_baseline].mean()
spec_godec -= spec_godec[is_baseline].mean()

std_psw = spec_psw[is_baseline].std()
std_godec = spec_godec[is_baseline].std()
```

```python
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(4.5, 4.5), dpi=300)

ax = axes[0]
ax.step(f, spec_psw, lw=1, label="CW + baseline subtraction")
ax.plot(f_fine, spec_fine, lw=2, alpha=0.25, color="black", zorder=-1, label="Ground truth")
ax.axhspan(-std_psw, std_psw, alpha=0.5, fc="orange", ec=None, zorder=-2, label="$\pm 1 \sigma$ noise level")
ax.set_ylabel("Brightness (arbitrary unit)")

ax = axes[1]
ax.step(f, spec_godec, lw=1, label="GoDec for fast-sampled PSW")
ax.step(f_fine, spec_fine, lw=2, alpha=0.25, color="black", zorder=-1, label="Ground truth")
ax.axhspan(-std_godec, std_godec, alpha=0.5, fc="orange", ec=None, zorder=-2, label="$\pm 1 \sigma$ noise level")
ax.set_xlabel("Frequency (arbitrary unit)")
ax.set_ylabel("Brightness (arbitrary unit)")

for ax in axes:
    ax.set_xlim(f.min(), f.max())
    ax.set_ylim(-0.04, 0.25)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.legend(loc="upper left", frameon=False)

fig.tight_layout()
savefig_in_notebook(fig, "mock-simulation.pdf")
```
